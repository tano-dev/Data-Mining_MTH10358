```python
df2 = standard_df.copy()

df2['equi-width_column0'] = pd.cut(x=df2.iloc[:, 0], bins=10) 
print('Roi rac hoa cot 0 bang 10 equi-width ranges:\n', df2)
```

### 2. Bài tập vận dụng

```python
link_hd_ko_che= r'https://archive.ics.uci.edu/ml/machine-learning-databases/arrhythmia/arrhythmia.data'
df = pd.read_csv(link_hd_ko_che, header=None)
df
```

```python
df.info()
df.describe()
print(df.dtypes)
```

```python
print(df.isnull().sum())
df = df.dropna()

df
```

```python

print(df.duplicated().sum())
df = df.drop_duplicates()
df
```

```python
(df== '?').any().value_counts()

df.replace('?', 0, inplace=True)
```

```python
scaler = StandardScaler()
array = df.values

robust_array = scaler.fit_transform(array)
robust_df = pd.DataFrame(robust_array, columns=df.columns)

print('Loai bo cac outliers: \n', robust_df)
```

```python
scaler = StandardScaler()
array = df.values

standard = scaler.fit_transform(array)
standard_df = pd.DataFrame(standard, columns=df.columns)
print('Chuan hoa du lieu:\n', standard_df)
```

```python
scaler = MinMaxScaler()
minmax = scaler.fit_transform(array)
minmax_df = pd.DataFrame(minmax, index = df.index)
print('Scaling dữ liệu:\n',minmax_df)
```

```python

zeros_cols = df.columns[(df == 0).all()]
df.drop(labels=zeros_cols, inplace=True, axis=1)

equi_width_scaler = KBinsDiscretizer(
    n_bins=10, encode='ordinal', strategy='uniform')
equi_width_data = equi_width_scaler.fit_transform(df)
equi_width_data
```

```python
equi_depth_scaler = KBinsDiscretizer(
    n_bins=10, encode='ordinal', strategy='quantile')
equi_depth_data = equi_width_scaler.fit_transform(df)
equi_depth_data
```
